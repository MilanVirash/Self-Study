linear regression model :

- Simple Linear Regression is a basic form of regression analysis.​​
- It involves one input variable (feature) and one output variable (target).
- The relationship is expressed by the equation: y = mx + b.​
- y: Target variable (what we want to predict).​
- x: Feature or input variable (what we use to make predictions).​
- m: Slope of the line (represents the relationship’s steepness).​
- b: Intercept (the point where the line intersects the y-axis).​

Steps :

- Data Collection: Gather input features and target variable.​
- Data Preprocessing: Handle missing data, outliers, and scaling.​
- Data Splitting: Divide into training and testing sets.​
- Model Selection: Choose a regression technique (e.g., linear).​
- Model Training: Fit the model to the training data.​
- Model Evaluation: Assess performance using metrics (MSE, R2).​
- Model Interpretation: Analyze feature importance.​
- Prediction: Use the model for new data.​

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# generate 2 colums data for example

np.random.seed(0)
hours_studied = np.random.uniform(0, 10, 100)
exam_scores = 50 + 5 * hours_studied + np.random.normal(0, 2, 100)

# create dataframe

df = pd.DataFrame({"Hours_Studied": hours_studied, "Exam_Scores": exam_scores})
df
```

```python
X = df[["Hours_Studied"]]
Y = df["Exam_Scores"]
```

```python
# train test split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# model fit

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# predict the values

y_pred = model.predict(x_test)
y_pred
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"Mean Squared Error : {mse:.2f}")
print(f"R2 Score : {r2:.2f}")
print(f"Mean Absolute Error : {mae:.2f}")
```

```python
# Plot the regression line and data points
plt.figure(figsize=(8, 5), facecolor='#EF9481')
plt.scatter(x_test, y_test, color='black', label='Actual Data')
plt.plot(x_test, y_pred, color='yellow', label='Regression Line', linewidth=2)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.legend()
plt.title('Simple Linear Regression: Hours Studied vs. Exam Scores')
plt.grid(True)
plt.show()
```
