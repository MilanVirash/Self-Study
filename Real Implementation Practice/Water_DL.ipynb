#  <b> • Outlier Treatment:

<div style="border-radius:10px; padding: 15px; background-color: #975455; font-size:115%; text-align:left; width: 90%;">
<p align="left"><font color="White" style="font-family: 'Book Antiqua';">
In a water potability dataset, outliers could represent instances where water quality is dangerously high or low for certain contaminants. These extreme values could be crucial for identifying water sources that pose significant health risks.

Removing outliers can distort the natural distribution of the data, potentially leading to misleading statistical summaries or modeling results.

Some machine learning models, especially those based on decision trees (like Random Forests) or those using robust methods (like SVM with an appropriate kernel), can handle outliers effectively. Removing outliers might deprive these models of valuable information.

And our dataset is already small, if we remove the outliers, it will get even smaller.

So I did not remove OUTLIERS but replace it with lower/upper whiskers.
</font></p>

```python
# Function to replace outliers with lower/upper whiskers
def replace_outliers(data):
    for col in data.columns:
        if data[col].dtype != 'object':  # Apply only to numeric columns
            Q1 = data[col].quantile(0.25)
            Q3 = data[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_whisker = Q1 - 1.5 * IQR
            upper_whisker = Q3 + 1.5 * IQR
            
            # Replace values below the lower whisker
            data[col] = data[col].apply(lambda x: lower_whisker if x < lower_whisker else x)
            # Replace values above the upper whisker
            data[col] = data[col].apply(lambda x: upper_whisker if x > upper_whisker else x)
    return data

# Apply the function to the dataset
cleaned_data = replace_outliers(data)

# Check the result
cleaned_data.describe()
```

#  <b> • Data Splitting And SMOTE For make the Data Balanced:

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

```python
x = data.drop('Potability', axis=1)
y = data['Potability']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=24)
```

```python
from sklearn.mixture import GaussianMixture

# Fit GMM on the minority class (Potability == 1) to generate synthetic data
gmm = GaussianMixture(n_components=2, random_state=42)

# Train the GMM on the minority class data
X_minority = x_train[y_train == 1]
gmm.fit(X_minority)

# Generate synthetic samples from GMM
synthetic_data, _ = gmm.sample(n_samples=100)  # Specify the number of synthetic samples

# Create a DataFrame for the synthetic samples
synthetic_df = pd.DataFrame(synthetic_data, columns=x.columns)

# Assign the target label (Potability == 1) to the synthetic data
synthetic_labels = np.ones(synthetic_df.shape[0])

# Combine original training data with synthetic data
X_train_Smote = np.vstack([x_train, synthetic_df])
y_train_Smote = np.hstack([y_train, synthetic_labels])
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create a baseline neural network model
def create_baseline():
    model = Sequential()
    model.add(Dense(60, input_dim=x_train.shape[1], activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# Initialize the model
baseline_model = create_baseline()

# Train the model
baseline_model.fit(x_train, y_train_Smote, epochs=100, batch_size=5, verbose=0)

# Evaluate the model on the test set
loss, accuracy = baseline_model.evaluate(x_test, y_train_Smote)
print(f"Baseline Model Accuracy: {accuracy:.2f}")
```
