For our __Exploratory Data Analysis (EDA)__, we'll take it in two main steps:

* __1. Univariate Analysis:__:
    - Here, we'll focus on one feature at a time to understand its distribution and range.
    
    
* __2. Bivariate Analysis:__:
    - In this step, we'll explore the relationship between each feature and the target variable. This helps us figure out the importance and influence of each feature on the target outcome.

With these two steps, we aim to gain insights into the individual characteristics of the data and also how each feature relates to our main goal: __predicting the target variable__.

<a id="basic"></a>
# <b><span style='color:'>Step 4.1 |</span><span style='color:blue'> Univariate Analysis</span></b>

```python
# Get continuous columns from your DataFrame
continuous_features = df.select_dtypes(include=[np.number]).columns

# Set up the subplot grid
n_features = len(continuous_features)
n_rows = (n_features + 2) // 3  # Calculate the number of rows needed
fig, ax = plt.subplots(nrows=n_rows, ncols=3, figsize=(18, n_rows * 6))

# Flatten the axes array for easy indexing
ax = ax.flatten()

# Loop to plot histograms for each continuous feature
for i, col in enumerate(continuous_features):
    values, bin_edges = np.histogram(df[col].dropna(), bins=15)  # Set a fixed bin size for uniformity
    
    graph = sns.histplot(data=df, x=col, bins=bin_edges, kde=True, ax=ax[i],
                         edgecolor='none', color='#ff826e', alpha=0.7, line_kws={'lw': 2})
    ax[i].set_xlabel(col, fontsize=16)
    ax[i].set_ylabel('Count', fontsize=16)
    ax[i].set_xticks(np.round(bin_edges, 1))
    ax[i].set_xticklabels(ax[i].get_xticks(), rotation=45, fontsize=14)
    ax[i].set_yticklabels(ax[i].get_yticks(), fontsize=14)
    ax[i].grid(color='lightgrey', linestyle='--', alpha=0.7)

    max_height = max(values)
    for p in graph.patches:
        if p.get_height() > 0:
            ax[i].annotate(f'{int(p.get_height())}', 
                           (p.get_x() + p.get_width() / 2, p.get_height() + max_height * 0.02),
                           ha='center', fontsize=12, fontweight="bold", color='black')
    
    textstr = '\n'.join((
        r'$\mu=%.2f$' % df[col].mean(),
        r'$\sigma=%.2f$' % df[col].std()
    ))
    ax[i].text(0.75, 0.85, textstr, transform=ax[i].transAxes, fontsize=14, verticalalignment='top',
               color='black', bbox=dict(boxstyle='round', facecolor='#ff826e', edgecolor='white', pad=0.6))

# Turn off any unused subplots
for i in range(n_features, len(ax)):
    ax[i].axis('off')

plt.suptitle('Distribution of Continuous Variables', fontsize=24, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to fit suptitle
plt.show()
```

    <p style="background-color:blue; font-family:Book Antiqua; color:white; font-size:230%; text-align:left; border-radius:0px 0px;">• Inference</p>
    <p style="font-family: 'Book Antiqua'; color: white; font-size:130%">

In summary, outliers in the METABRIC breast cancer dataset may contain significant clinical information that could enhance the robustness of predictive models.

Removing these outliers without careful consideration might result in the loss of crucial insights and potentially reduce model performance.

Rather than removing outliers, it may be more appropriate to apply techniques such as data transformation or robust modeling methods to address outliers while preserving their informative value.

</p>
</div>


```python
# Function to remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply outlier removal
for column in numeric_columns:
    df = remove_outliers(df, column)
```

    <p style="background-color:blue; font-family:Book Antiqua; color:white; font-size:230%; text-align:left; border-radius:0px 0px;">• Inference</p>
    <p style="font-family: 'Book Antiqua'; color: white; font-size:130%">

The provided code uses the Interquartile Range (IQR) method to identify and remove outliers from the dataset. While this approach effectively eliminates extreme values that could skew analysis, it may not be ideal for the METABRIC breast cancer dataset.

Outliers in medical datasets often represent rare and clinically significant cases that could offer valuable insights. Removing these outliers without careful consideration could lead to the loss of important information and potentially decrease the robustness of predictive models.

Therefore, while the code is technically sound for outlier removal, it might be more appropriate to consider alternative methods like data transformation or robust modeling techniques. These approaches allow you to address the influence of outliers while preserving their informative value within the dataset.

# <b><span style='color:blue'>Step 5.4 |</span><span style='color:white'> Transforming Skewed Features </span></b>

<div style="border-radius:10px; padding: 15px; background-color: blue; font-size:115%; text-align:left; width: 70%;">
<h3 align="left"><font color=white>To avoid data leakage and ensure our model generalizes well to unseen data:</font></h3>

__1- Data Splitting:__ We'll first split our dataset into a training set and a test set. This ensures that we have a separate set of data to evaluate our model's performance, untouched during the training and preprocessing phases.

__2- Applying Transformations to Test Data:__ Once our transformation parameters are determined from the training set, we'll use these exact parameters to transform our validation/test set. This approach ensures that no information from the validation/test set leaks into our training process.

__3. Hyperparameter Tuning & Cross-Validation:__ Given our dataset's size, to make the most of the available data during the model training phase, we'll employ __cross-validation on the training set for hyperparameter tuning__. This allows us to get a better sense of how our model might perform on unseen data, without actually using the test set. The test set remains untouched during this phase and is only used to evaluate the final model's performance.


By following this structured approach, we ensure a rigorous training process, minimize the risk of data leakage, and set ourselves up to get a realistic measure of our model's performance on unseen data.
