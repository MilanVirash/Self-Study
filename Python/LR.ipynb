```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
HourStudies = np.random.uniform(0, 10, 1000)
ExamScore =  50 + 5*HourStudies + np.random.normal(0, 2, 1000)
```

```python
df = pd.DataFrame({"HourStudies":HourStudies, "ExamScore":ExamScore})
df
```

```python
x = df[['HourStudies']]
y = df['ExamScore']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

MSE = mean_squared_error(y_test, y_pred)
R2 = r2_score(y_test, y_pred)

print(f'MSE : {MSE:.2f}')
print(f'R2 : {R2:.2f}')
```

```python
import matplotlib.pyplot as plt

# Plot the regression line and data points
plt.figure(figsize=(8, 5), facecolor='#EF9481')
plt.scatter(x_test, y_test, color='black', label='Actual Data')
plt.plot(x_test, y_pred, color='yellow', label='Regression Line', linewidth=2)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.legend()
plt.title('Simple Linear Regression: Hours Studied vs. Exam Scores')
plt.grid(True)
plt.show()
```
